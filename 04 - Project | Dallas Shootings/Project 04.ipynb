```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect('/content/sample_data/dallas-ois.sqlite')
```

```python
incidents = pd.read_sql('SELECT * FROM incidents', conn)
officers = pd.read_sql('SELECT * FROM officers', conn)
subjects = pd.read_sql('SELECT * FROM subjects', conn)
```

```python
data = pd.read_sql('SELECT * FROM subjects JOIN incidents ON subjects.case_number=incidents.case_number', conn)
```

```python
dataset = data[['race', 'subject_statuses', 'latitude', 'longitude']].dropna()
```

```python
dataset.head()
```

```python
len(dataset)
```

```python
mapping = {'B': 'b', 'W': 'y', 'A': 'r', 'L': 'c'}
dataset['race'] = data['race'].apply(lambda x: mapping[x])
dataset.head()
```

```python
mapping = {'Deceased': 1000, 'Injured': 500, 'Shoot and Miss': 250}
dataset['subject_statuses'] = dataset['subject_statuses'].apply(lambda x:mapping.get(x, 100))
```

```python
dataset.plot.scatter(x='longitude', y='latitude', s='subject_statuses', c='race', alpha=.25)
```

```python
dataset.plot.scatter(x='longitude', y='latitude', s='subject_statuses', c='race', alpha=.25, title='Dallas Shootings')
```

```python
import folium
```

```python
m = folium.Map(location=[32.8, -96.8])
```

```python
m
```

```python
for _, row in dataset.iterrows():
  folium.CircleMarker(
      location=[row['latitude'], row['longitude']],
      radius=row['subject_statuses']//100,
      color=row['race'],
      fill=True,
      fill_color=row['race']
  ).add_to(m)
```
