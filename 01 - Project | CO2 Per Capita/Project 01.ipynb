```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://github.com/ad3rinto/data-science-rune/raw/refs/heads/main/files/WorldBank-ATM.CO2E.PC_DS2.csv', index_col=0)
```

```python
data.isnull().any()
```

```python
data.isnull().sum().plot.hist(figsize=(20,6))
```

```python
data_clean = data.dropna(axis='columns')
```

```python
len(data.columns), len(data_clean.columns)
```

```python
data_plot = (data_clean.loc[2018]-data_clean.loc[1998])/data_clean.loc[1998]
```

```python
data_plot.describe()
```

```python
data_plot.plot.hist(bins=10)
```

```python
(data_plot < 0).value_counts()
```

```python
(data_plot < 0).value_counts().plot.pie(title='Countries with decrease CO2', colors=['r', 'g'], labels=['>=0', '< 0'], ylabel='1998-2018', autopct='%1.1f%%')
```

```python
data_plot2 = (data_clean.loc[2018]-data_clean.loc[2008])/data_clean.loc[2008]
```

```python
data_plot2.describe()
```

```python
data_plot2.plot.hist()
```

```python
(data_plot2 < 0).value_counts()
```

```python
(data_plot2 < 0).value_counts().plot.pie(title='Countries with decrease CO2 (2008-2018)', colors=['r', 'g'], labels=['>=0', '< 0'], ylabel='2008-2018', autopct='%1.1f%%')
```
