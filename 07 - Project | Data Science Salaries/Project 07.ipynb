Report

```python
plot_data = data.groupby(['company_size', 'experience_level'])['salary_in_usd'].mean().unstack()
plot_data
```

```python
plot_data = plot_data.reindex(index=['S','M','L'])
plot_data = plot_data[['EN','MI','SE','EX']]
plot_data
```

```python
plot_data.plot.bar(ylabel='USD salary')
```
