```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://github.com/ad3rinto/data-science-rune/raw/refs/heads/main/files/API_SP/API_SP.POP.TOTL_DS2_en_csv_v2_3158886.csv', skiprows=4)
data.head()
```

```python
data.index
```

```python
data[data['Country Name'] == 'World']
```

```python
data.dtypes
```

```python
data.drop(['Country Code', 'Indicator Code'], axis='columns')
```

```python
data.dropna(how='all', axis='columns')
```

```python
dataset = data[data['Country Name'] == 'World'].drop(['Country Name', 'Country Code', 'Indicator Code', 'Indicator Name'], axis='columns').dropna(how='all', axis='columns')
dataset
```

```python
dataset.transpose()
```

```python
dataset = dataset.transpose().rename(columns={259:'World Population'})
dataset.head()
```

```python
dataset.plot()
```

```python
dataset['Yearly Growth'] = dataset['World Population'].pct_change()
dataset.head()
```

```python
dataset['Yearly Growth'].plot(title='Yearly Growth', ylabel='in %', ylim=0)
```

```python
dataset['Smooth'] = dataset['Yearly Growth'].rolling(10).mean()
dataset[['Smooth', 'Yearly Growth']].plot()
```

```python
dataset[['Smooth', 'Yearly Growth']]*=100
```

```python
dataset[['Smooth', 'Yearly Growth']].plot(ylabel='Yearly Growth (%)', xlabel='Year', title='World Population Growth', ylim=0)
```
